```python
import pandas as pd
import json
import numpy as np
```

```python
global_hunger = pd.read_csv('global-hunger-index.csv')
uw = pd.read_csv('share-of-children-underweight.csv')
wasting = pd.read_csv('share-of-children-with-a-weight-too-low-for-their-height-wasting.csv')
stunt = pd.read_csv('share-of-children-younger-than-5-who-suffer-from-stunting.csv')

pop = pd.read_csv('pop1.csv')
gdp = pd.read_csv('GDP by Country 1999-2022.csv')
```

```python
with open('world.json', 'r') as f:
  data = json.load(f)

```

```python
for countries in data['objects']['countries1']['geometries']:
    name = countries['properties']['name']
    years = ['2000','2006','2012','2021']

    if name in global_hunger.Entity.unique():
        countries['properties']['global_hunger'] = {}
        for year in years:
            if len(global_hunger[(global_hunger.Entity == name) & (global_hunger.Year == int(year))].iloc[:, 3])>0:
                countries['properties']['global_hunger'][year] = global_hunger[(global_hunger.Entity == name) & (global_hunger.Year == int(year))].iloc[:, 3].values[0]
```

```python
years = uw.Year.unique()
print(sorted(years))
```

```python
for countries in data['objects']['countries1']['geometries']:
    name = countries['properties']['name']

    if name in uw.Entity.unique():
        years = sorted(uw.Year.unique())
        countries['properties']['underweight'] = {}
        for year in years:
            if len(uw[(uw.Entity == name) & (uw.Year == year)].iloc[:, 3])>0:
                countries['properties']['underweight'][str(year)] = uw[(uw.Entity == name) & (uw.Year == year)].iloc[:, 3].values[0]
    if name in wasting.Entity.unique():
        years = sorted(wasting.Year.unique())
        countries['properties']['wasting'] = {}
        for year in years:
            if len(wasting[(wasting.Entity == name) & (wasting.Year == year)].iloc[:, 3])>0:
                countries['properties']['wasting'][str(year)] = wasting[(wasting.Entity == name) & (wasting.Year == year)].iloc[:, 3].values[0]
    if name in stunt.Entity.unique():
        years = sorted(stunt.Year.unique())
        countries['properties']['stunt'] = {}
        for year in years:
            if len(stunt[(stunt.Entity == name) & (stunt.Year == year)].iloc[:, 3])>0:
                countries['properties']['stunt'][str(year)] = stunt[(stunt.Entity == name) & (stunt.Year == year)].iloc[:, 3].values[0]
```

```python
years = gdp.columns.values[1:]
print(years)
```

```python
for countries in data['objects']['countries1']['geometries']:
    years = gdp.columns.values[1:]
    name = countries['properties']['name']

    if name in gdp.Country.unique():
        countries['properties']['gdp'] = {}
        for year in years:
            gdpvalue = float(gdp[gdp.Country==name][str(year)].values[0].replace(',',''))
            if gdpvalue > 0:
                countries['properties']['gdp'][str(year)] = gdpvalue
    
```

```python
for countries in data['objects']['countries1']['geometries']:
    print(countries)
    print("\n")
```

```python
with open('data.json', 'w') as f:
    json.dump(data, f)
```
